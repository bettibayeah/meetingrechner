# 3. Meetingkosten

##### 3.1) Zusammentragen der Stundenlöhne
Aus dem Quartalbericht des statistischen Bundesamt ergibt sich, dass Voll- und Teilzeitbeschäftigte im Dienstleistungsbereich (also z.B. wirtschaftliche Dienstleistungen, Gastgewerbe, Grundstücks- und Wohnungswesen) 2021 im Schnitt
27,14 Euro pro Stunde verdient haben.
Nach Leistungsgruppen geordnet, verdienen leitende Angestellte 59 Euro (LG1), herausgehobene Fachkräfte 36 Euro (LG2), Fachkräfte 22 Euro (LG3; es geht bis LG5).

_Der genaue Datensatz findet sich unter: Arbeitnehmerverdienste - Fachserie 16 Reihe 2.1 - 4. Vierteljahr 2021 (Letzte Ausgabe - berichtsweise eingestellt_


Man könnte so unterteilen: LG1 = Vorstand / Geschäftsführung; LG2 = Abteilungsleitung / Senior Management und Teamleitung / Middle Management; LG3 = Fachkraft / Mitarbeiter:in

```python
hwage_vorstand = 59
hwage_leitung = 36
hwage_mitarbeiter = 22

minwage_vorstand = 59/60
minwage_leitung = 36/60
minwage_mitarbeiter = 22/60

meeting_effectiveness_list = [
    ("Wie viele Personen aus der Geschäftsführung waren anwesend?" , hwage_vorstand, minwage_vorstand),
    ("Wie viele Personen aus der Teamleitung waren anwesend?", hwage_leitung, minwage_leitung),
    ("Wie viele weitere Mitarbeitende waren anwesend?", hwage_mitarbeiter, minwage_mitarbeiter)
]

# Markdown-Tabelle erstellen
markdown_table = "| Frage | Kosten pro Person pro Stunde | pro Minute |\n|--------|-----------------------------------|------------------------|\n"
for faktor, h, min in meeting_effectiveness_list:
    markdown_table += f"| {faktor} | {h} Euro | {min} Euro |\n"

print(markdown_table)
```

| Frage | Kosten pro Person pro Stunde | pro Minute |
|--------|-----------------------------------|------------------------|
| Wie viele Personen aus der Geschäftsführung waren anwesend? | 59 Euro | 0.9833333333333333 Euro |
| Wie viele Personen aus der Teamleitung waren anwesend? | 36 Euro | 0.6 Euro |
| Wie viele weitere Mitarbeitende waren anwesend? | 22 Euro | 0.36666666666666664 Euro |


```python
min = 50
vorstand = 2
leitung = 3
mitarbeiter = 10

def meetingkosten(min, minwage_vorstand, minwage_leitung, minwage_mitarbeiter, vorstand, leitung, mitarbeiter):
    kosten = min * (minwage_vorstand * vorstand + minwage_leitung * leitung + minwage_mitarbeiter * mitarbeiter)
    return kosten

print(meetingkosten(min, minwage_vorstand, minwage_leitung, minwage_mitarbeiter, vorstand, leitung, mitarbeiter))

```

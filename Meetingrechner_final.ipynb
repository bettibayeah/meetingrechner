### B) Die Kostenrechnung

Die Datengrundlage 

```python
hwage_vorstand = 59
hwage_leitung = 36
hwage_mitarbeiter = 22

minwage_vorstand = 59/60
minwage_leitung = 36/60
minwage_mitarbeiter = 22/60

meeting_effectiveness_list = [
    ("Wie viele Personen aus der Geschäftsführung waren anwesend?" , hwage_vorstand, minwage_vorstand),
    ("Wie viele Personen aus der Teamleitung waren anwesend?", hwage_leitung, minwage_leitung),
    ("Wie viele weitere Mitarbeitende waren anwesend?", hwage_mitarbeiter, minwage_mitarbeiter)
]


markdown_table = "| Frage | Kosten pro Person pro Stunde | pro Minute |\n|--------|-----------------------------------|------------------------|\n"
for faktor, h, min in meeting_effectiveness_list:
    markdown_table += f"| {faktor} | {h} Euro | {min} Euro |\n"

print(markdown_table)
```

```python

def meetingkosten(min, minwage_vorstand, minwage_leitung, minwage_mitarbeiter, vorstand, leitung, mitarbeiter):
    kosten = min * (minwage_vorstand * vorstand + minwage_leitung * leitung + minwage_mitarbeiter * mitarbeiter)
    return kosten

print(meetingkosten(min, minwage_vorstand, minwage_leitung, minwage_mitarbeiter, vorstand, leitung, mitarbeiter))

```
